# ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO LA REGRESIÓN LOGÍSTICA

## Modelo de regrsión logistica

```python
from sklearn.model_selection import train_test_split
```

```python
# Dividir conjunto de entrenamiento 75% para crear el modelo y el 25% para probarlo 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Importar las librerias para entrenar el modelo y para evaluar uno de los KPI - accuracy

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Entrenamos el modelo de regresión logistica

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predecimos en base al 25% d elos datos apra probarlo

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
```

```python
print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test)))
print("-----------------------------------------------------------")
print(classification_report(y_test, y_pred))
```

Como resultado tenemos:

* Un Accuracy del 87.5%, este nos indica que está prediciendo el 87.5% de los resultados
* Para aquellos que no abandonaran la empresa tuvo una precisión del 88%, un recall del 98% y un f1-score del 93%, es decir que predijo con una gran tasa de acierto aquellos que no abandonarían la empresa
* Para aquellos que abandonaran la empresa tuvo una precisión del 78%, un recall del 34% y un f1-score del 48%, es decir que predijo con una baja tasa de acierto aquellos que abandonarían la empresa
* En conclusión, no es útil el modelo dado que, si bien tuvo una buena tasa en cuanto a la predicción de aquellos que no abandonarían la empresa, no predijo aquellos que abandonarían y es este el que eremos saber
